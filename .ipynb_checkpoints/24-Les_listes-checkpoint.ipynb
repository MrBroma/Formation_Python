### Ajouter ou supprimer des éléments à une liste

Pour ajouter un élément on utilise la méthode append

```python
# Ajout d'un élément à la fois
```

```python
liste.append(6)
liste
```

```python
# Ajout de plusieurs éléments avec la méthose .extend()
```

```python
liste.extend([7, 8, 9])
print(liste)
```

```python
# Supprimer un élément d'un liste --> première occurence
liste.remove(5)
print(liste)
```
